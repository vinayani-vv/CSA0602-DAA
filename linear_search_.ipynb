```python
n = int(input("Enter number of elements: "))
arr = []
for i in range(n):
    arr.append(int(input("Enter element: ")))
key = int(input("Enter key to search: "))
found = False
for i in range(len(arr)):
    if arr[i] == key:
        print("Key found at index", i)
        found = True
        break
if not found:
    print("Key not found")
```
